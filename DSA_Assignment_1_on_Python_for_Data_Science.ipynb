```python
irisData[['SL','SW','PL','PW']].plot.hist()   # Histogram
```

```python
irisData[['SL','SW','PL','PW']].plot.kde() #Kernel based density estimation which visualises the probability density at different values of a continuous variable.
                                           # Here, the variables used are SL,SW,PL and PW
```

```python
sns.pairplot(irisData)#, hue = 'Classification')      # Use of pair plot which allows comparison between all the available numerical columns in the data
```

```python
slice_data=irisData[['SL','SW','PL']]
sns.pairplot(slice_data)                              #Allows pair plot with selected data
```

```python
sns.heatmap(slice_data.corr(),annot=True)      
plt.xticks(rotation=45)  # Command to adjust the slanting position of xlabels  
# plotting the extend of correleation between sliced variables
```

```python

sns.violinplot(x = 'SW', y = 'Classification', data = irisData, inner = 'quartile')  #violinploit
```

```python
sns.stripplot(x = 'Classification', y = 'PW', data = irisData, jitter=True)
```
